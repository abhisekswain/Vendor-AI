```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt


print (cv2.__version__)

# I cropped out each stereo image into its own file.
# You'll have to download the images to run this for yourself
chanel1 = cv2.imread('chanel1.jpeg')
chanel2 = cv2.imread('chanel2.jpeg')

def show_rgb_img(img):
    """Convenience function to display a typical color image"""
    return plt.imshow(cv2.cvtColor(img, cv2.CV_32S))
```

```python
show_rgb_img(chanel1)
```

```python
show_rgb_img(chanel2)
```

```python
def to_gray(color_img):
    gray = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)
    return gray

chanel1_gray = to_gray(chanel1)
chanel2_gray = to_gray(chanel2)

plt.imshow(chanel1_gray, cmap='gray')
```

```python
plt.imshow(chanel2_gray, cmap='gray')
```

```python
def gen_sift_features(gray_img):
    sift = cv2.xfeatures2d.SIFT_create()
    # kp is the keypoints
    #
    # desc is the SIFT descriptors, they're 128-dimensional vectors
    # that we can use for our final features
    kp, desc = sift.detectAndCompute(gray_img, None)
    return kp, desc

def show_sift_features(gray_img, color_img, kp):
    return plt.imshow(cv2.drawKeypoints(gray_img, kp, color_img.copy()))

# generate SIFT keypoints and descriptors
chanel1_kp, chanel1_desc = gen_sift_features(chanel1_gray)
chanel2_kp, chanel2_desc = gen_sift_features(chanel2_gray)


show_sift_features(chanel1_gray, chanel1, chanel1_kp);
```

```python
show_sift_features(chanel2_gray, chanel2, chanel2_kp);
```

```python
# create a BFMatcher object which will match up the SIFT features
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)

matches = bf.match(chanel1_desc, chanel2_desc)

# Sort the matches in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)

# draw the top N matches
N_MATCHES = 100

match_img = cv2.drawMatches(
    chanel1, chanel1_kp,
    chanel2, chanel2_kp,
    matches[:N_MATCHES], chanel2.copy(), flags=0)

plt.figure(figsize=(12,6))
plt.imshow(match_img);
```

```python
def explain_keypoint(kp):
    print ('angle\n', kp.angle)
    print ('\nclass_id\n', kp.class_id)
    print ('\noctave (image scale where feature is strongest)\n', kp.octave)
    print ('\npt (x,y)\n', kp.pt)
    print ('\nresponse\n', kp.response)
    print ('\nsize\n', kp.size)

print ('this is an example of a single SIFT keypoint:\n* * *')
explain_keypoint(chanel1_kp[0])
```
